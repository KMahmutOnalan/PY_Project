```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions and initial color
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Gamma Correction Measurement')

# Define colors
colors = {
    'white': (255, 255, 255),
    'red': (255, 0, 0),
    'green': (0, 255, 0),
    'blue': (0, 0, 255)
}

# Color keys list for easy navigation
color_keys = list(colors.keys())

# Initial settings
color_index = 0
luminance = 0
max_luminance = 255
step = 15

def set_color_luminance(color, luminance):
    r, g, b = colors[color]
    r = int(r * (luminance / 255))
    g = int(g * (luminance / 255))
    b = int(b * (luminance / 255))
    return (r, g, b)

# Main loop
running = True
while running:
    # Fill the screen with the current color and luminance
    color_name = color_keys[color_index]
    screen.fill(set_color_luminance(color_name, luminance))
    
    # Update the display
    pygame.display.flip()
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False
            elif event.key == pygame.K_UP:
                luminance = min(max_luminance, luminance + step)
            elif event.key == pygame.K_DOWN:
                luminance = max(0, luminance - step)
            elif event.key == pygame.K_RIGHT:
                color_index = (color_index + 1) % len(color_keys)
            elif event.key == pygame.K_LEFT:
                color_index = (color_index - 1) % len(color_keys)
    
    # Print the current color and luminance
    print(f'Color: {color_name}, Luminance: {luminance}')
    
pygame.quit()
sys.exit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Gamma Correction Measurement')

# Initial settings
square_size = 200
square_x = (screen_width - square_size) // 2
square_y = (screen_height - square_size) // 2
luminance = 0
max_luminance = 255
step = 15

# List to store brightness values for each trial
brightness_values = []

# Function to set luminance for a gray square
def set_square_luminance(luminance):
    return (luminance, luminance, luminance)

# Main loop
running = True
advance_to_next_trial = False

while running:
    # Fill the screen with black
    screen.fill((0, 0, 0))
    
    # Draw the square with the current luminance
    luminance_color = set_square_luminance(luminance)
    pygame.draw.rect(screen, luminance_color, (square_x, square_y, square_size, square_size))
    
    # Update the display
    pygame.display.flip()
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_q:  # Use Q key to quit
                running = False
            elif event.key == pygame.K_UP:
                luminance = min(max_luminance, luminance + step)
            elif event.key == pygame.K_DOWN:
                luminance = max(0, luminance - step)
            elif event.key == pygame.K_SPACE:
                advance_to_next_trial = True
    
    # Print the current luminance
    print(f'Luminance: {luminance}')
    
    # Advance to the next trial
    if advance_to_next_trial:
        # Save the current luminance value
        brightness_values.append(luminance)
        
        # Reset the luminance for the next trial
        luminance = 0
        advance_to_next_trial = False

pygame.quit()

# Print the saved brightness values
#print('Brightness values for each trial:', brightness_values)
sys.exit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Gamma Correction Measurement')

# Initial settings
square_size = 200
square_x = (screen_width - square_size) // 2
square_y = (screen_height - square_size) // 2
max_luminance = 255
luminance = max_luminance  # Set initial luminance to the brightest
step = 15

# List to store brightness values for each trial
brightness_values = []

# Function to set luminance for a gray background
def set_background_luminance(luminance):
    return (luminance, luminance, luminance)

# Main loop
running = True
advance_to_next_trial = False

while running:
    # Fill the screen with the current luminance
    luminance_color = set_background_luminance(luminance)
    screen.fill(luminance_color)
    
    # Draw the square in the center with the current luminance (optional)
    pygame.draw.rect(screen, (0, 0, 0), (square_x, square_y, square_size, square_size))  # Black square
    
    # Update the display
    pygame.display.flip()
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_q:  # Use Q key to quit
                running = False
            elif event.key == pygame.K_UP:
                luminance = min(max_luminance, luminance + step)
            elif event.key == pygame.K_DOWN:
                luminance = max(0, luminance - step)
            elif event.key == pygame.K_SPACE:
                advance_to_next_trial = True
    
    # Print the current luminance
    print(f'Luminance: {luminance}')
    
    # Advance to the next trial
    if advance_to_next_trial:
        # Save the current luminance value
        brightness_values.append(luminance)
        
        # Reset the luminance for the next trial
        luminance = max_luminance  # Set the luminance to the brightest
        advance_to_next_trial = False

pygame.quit()

# Print the saved brightness values
print('Brightness values for each trial:', brightness_values)
sys.exit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up display
infoObject = pygame.display.Info()
screen = pygame.display.set_mode((infoObject.current_w, infoObject.current_h), pygame.FULLSCREEN)
pygame.display.set_caption('Gamma Correction Measurement')

# Colors
colors = {'white': (255, 255, 255), 'red': (255, 0, 0), 'green': (0, 255, 0), 'blue': (0, 0, 255)}
color_names = list(colors.keys())
current_color = 0
luminance = 0
max_luminance = 255

# Font
font = pygame.font.SysFont(None, 55)

# Main loop
running = True
while running:
    # Fill screen with current color and luminance
    color = colors[color_names[current_color]]
    scaled_color = (color[0] * luminance // max_luminance,
                    color[1] * luminance // max_luminance,
                    color[2] * luminance // max_luminance)
    screen.fill(scaled_color)

    # Render luminance value on screen
    luminance_text = font.render(f'Luminance: {luminance}', True, (255, 255, 255))
    screen.blit(luminance_text, (50, 50))

    # Update display
    pygame.display.flip()

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False
            elif event.key == pygame.K_UP:
                luminance = min(max_luminance, luminance + 1)
            elif event.key == pygame.K_DOWN:
                luminance = max(0, luminance - 1)
            elif event.key == pygame.K_RIGHT:
                current_color = (current_color + 1) % len(colors)
            elif event.key == pygame.K_LEFT:
                current_color = (current_color - 1) % len(colors)

# Quit Pygame
pygame.quit()
sys.exit()
```
